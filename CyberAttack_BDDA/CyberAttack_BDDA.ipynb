```python
def map_attacks_to_families(attack):
    """ Map attack to it's family """
    return attack_families[attack]


df["Labels"] = df["Labels"].apply(lambda attack: map_attacks_to_families(attack))
df1["Labels"] = df1["Labels"].apply(lambda attack: map_attacks_to_families(attack))
#df2["Labels"] = df2["Labels"].apply(lambda attack: map_attacks_to_families(attack))
print(pd.unique(df["Labels"]))
```

# 4.Data pipelining for data processing and modeling

```python
from imblearn.pipeline import Pipeline as imbpipeline
from imblearn.over_sampling import SMOTE
```

```python
# Which columns are numeric and which not
num_cols = ['Duration','Srcbytes','Dstbytes','Numfailedlogins','Suattempted',
            'Numroot','Numfilecreations','Numshells','Numaccessfiles',
            'Numoutboundcmds','Ishostlogin','Count','Rerrorrate','Srvrerrorrate']
cat_cols = ['Protocoltype']
```

```python
#Create pipe for both numerical as also categorical data
nc_pipe = imbpipeline([ ('impute', SimpleImputer(strategy = 'mean')), ('ss', StandardScaler()) ])
cat_pipe = imbpipeline([('impute', SimpleImputer(strategy = 'most_frequent')), ('ohe', OneHotEncoder())])
```

```python
# Test each pipe
nc_pipe.fit_transform(df[num_cols])
cat_pipe.fit_transform(df[cat_cols])
```

```python
# Now put the two pipes within one column-transformer
ct = ColumnTransformer([('nc', nc_pipe, num_cols), ('cp', cat_pipe,cat_cols)])
```

```python
# Test this column transformer pipe
ct.fit_transform(df)
```

```python
# We are using SMOTE because of our inbalanced datset
```

```python
# Final pipe to bring in the column-transformer followed by estimator
pt = imbpipeline([ ('ct', ct),
              ('sm',SMOTE( random_state=42)),('dt', DecisionTreeClassifier())])
```

```python
# Your train data
y = df.pop('Labels')
X = df
```

```python
# Split dataset. We will preprocess X_train and apply that
# processing to X_test later
X_train,X_test, y_train, y_test = train_test_split(
                                                    X,                   # Data features
                                                    y,                   # Target column
                                                    test_size = 0.2      # split-ratio
                                                    )
```

```python
f"X_train shape: {X_train.shape}"    
print()
f"X_test.shape : {X_test.shape}"     
print()
f"y_train shape: {y_train.shape}"    
print()
f"y_test shape : {y_test.shape}"     
```

```python
# Fit final pipe on the data
pt.fit(X_train,y_train)
```

```python
# Make predictions
pt.predict(X_test)
```

# 6. Hyperparameter tuning both using Grid Search and Randomized Search

```python
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RandomizedSearchCV
from sklearn.metrics import accuracy_score, f1_score
from sklearn.metrics import plot_roc_curve
```

```python
pt.get_params()
```

```python
parameters = {
              'dt__max_depth': [4,10],
              'dt__min_impurity_decrease':[0,1],
              'dt__min_samples_leaf': [2,3],
              'dt__criterion': ["gini", "entropy"]
              }                               # Total: 2 * 2 * 2 * 2 = 16
```

```python
grid_search = GridSearchCV(estimator=pt,           # pipeline object
                           param_grid=parameters,  # possible parameters
                           scoring = ['accuracy'], # Metrics for performance
                           cv=5,                   # No of folds
                           n_jobs=-1,              # USe parallel cpu threads
                           verbose =1,             # Higher the value, more the verbosity
                           refit = 'accuracy'      # Refitting final model on what parameters Those which maximise auc   
                           )
grid_search.fit(X_train, y_train)
cv_score = grid_search.best_score_
test_score = grid_search.score(X_test, y_test)
print(f'Cross-validation score: {cv_score}\nTest score: {test_score}')
```

```python
# Tune parameters using random search
# Create the object first
rs = RandomizedSearchCV(
                           estimator=pt,   # pipeline object
                           param_distributions=parameters, # possible parameters
                           scoring = ['accuracy'], #Metrics for performance
                           cv=5,               # No of folds
                           n_jobs=-1,          # USe parallel cpu threads 
                           verbose =1,         # Higher the value, more the verbosity
                           refit = 'accuracy'  # Refitting final model on what parameters Those which maximise auc   
                        )
rs.fit(X_train, y_train)
cv_score = rs.best_score_
test_score = rs.score(X_test, y_test)
print(f'Cross-validation score: {cv_score}\nTest score: {test_score}')
```
